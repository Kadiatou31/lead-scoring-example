```python
import pandas as pd
```

```python
data = pd.read_csv("dataset-example.csv")
```

```python
# Lead scoring
```

```python
def lead_score(row):
    score = 0

    # Taille d'entreprise
    if row['Taille_entreprise'] >= 100:
        score += 20
    elif row['Taille_entreprise'] >= 10:
        score += 10
    else:
        score += 5

    # Engagement
    score += row['Pages_visitees'] * 2
    score += row['Emails_ouverts'] * 3
    score += row['Appels_effectues'] * 5

    # Secteur prioritaire
    if row['Secteur'] in ['Tech', 'E-commerce']:
        score += 10

    return score
```

```python
# Apply the lead scoring function to each row
data['Score'] = data.apply(lead_score, axis=1)

# Sort the DataFrame by score
data_sorted = data.sort_values(by='Score', ascending=False)
print(data_sorted.head())
```
